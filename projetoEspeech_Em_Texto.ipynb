```python
from google.colab import files
from pydub import AudioSegment

# Função para fazer upload de arquivo
uploaded = files.upload()

# Exibir o nome do arquivo carregado
for file_name in uploaded.keys():
    print(f"Arquivo carregado: {file_name}")

# Carregar o arquivo OGG
audio_ogg = AudioSegment.from_ogg(file_name)  # Usar o nome do arquivo carregado

# Converter para WAV
audio_ogg.export("novo_arquivo.wav", format="wav")

print("Conversão concluída! Arquivo salvo como 'novo_arquivo.wav'.")

# Oferecer download do arquivo convertido
files.download("novo_arquivo.wav")
```

```python
# Inicializar o reconhecedor de fala
recognizer = sr.Recognizer()

# Abrir o arquivo de áudio carregado
audio_file = 'novo_arquivo.wav'  # Substitua pelo nome do arquivo carregado

with sr.AudioFile(audio_file) as source:
    audio_data = recognizer.record(source)
    texto = recognizer.recognize_google(audio_data, language='pt-BR')  # Pode mudar para 'en-US' para inglês
    print(f"Texto reconhecido: {texto}")
```
